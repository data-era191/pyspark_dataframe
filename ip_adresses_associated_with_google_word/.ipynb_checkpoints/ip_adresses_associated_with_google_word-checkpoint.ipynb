```python
from pyspark.sql import SparkSession
from pyspark.sql.functions import col,split
```

```python
spark=SparkSession.builder.appName("find_google_word_associated _with_ip_adresses").getOrCreate()
```

```python
df=spark.read.text("data.txt")
df.show(truncate=False)
```

```python
df = df.withColumn("ip", split(col("value"), " ")[0]) \
       .withColumn("date", split(col("value"), " ")[1]) \
       .withColumn("method",split(col("value")," ")[3]) \
       .withColumn("url",split(col("value")," ")[4]) \
       .drop("value")

df.show()
```

```python
google_ip=df.select("ip").where(col("url").contains("http://www.google.com"))
google_ip.show()
```

```python
google_ip.write.mode("overwrite").csv("./output",header=True)
```
